## 2. Dataset Exploration

```python
import pandas as pd
import json

# Load JSON
with open("/Users/felix/Documents/Data Science/05_The_project_plant_recognition/Data-Scientist-Project-Plant-Recognition-/01_data/v2-plant-seedlings-dataset-metadata.json") as f:
    metadata = json.load(f)

# Convert top-level fields to DataFrame
df_flat = pd.DataFrame([metadata])
print(df_flat[["name", "license", "datePublished"]])

df_flat.head()
```

```python
# Extract nested 'distribution' array
distributions = pd.json_normalize(metadata["distribution"])
print(distributions[["name", "contentSize", "encodingFormat"]])


distributions.head()
```

```python
# Example: Extract key metadata for a summary table
summary_data = {
    "Dataset Name": metadata["name"],
    "License": metadata["license"]["name"],
    "Classes": "12 (crop/weed seedlings)",
    "Image Count": "5,539",  # From description
    "Format": metadata["distribution"][1]["encodingFormat"]
}
df_summary = pd.DataFrame([summary_data])
print(df_summary)

df_summary.head()
```

```python
# Focus on key nested fields
df_license = pd.json_normalize(metadata["license"])
df_files = pd.json_normalize(metadata["distribution"])

# Combine into one DataFrame
df_combined = pd.concat([
    pd.DataFrame({k: [v] for k, v in metadata.items() 
                 if not isinstance(v, (dict, list))}),
    df_license.add_prefix("license_"),
    df_files.add_prefix("file_")
], axis=1)

df_combined.head()
```
